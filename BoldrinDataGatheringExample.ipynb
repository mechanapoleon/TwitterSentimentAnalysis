We start by importing a few useful libraries

```python
import tweepy
import pandas as pd
import numpy as np
```

Adding access data to the Twitter API

```python
consumer_key = "your key"
consumer_secret = "your consumer key"
access_token = "your access token"
access_secret = "your access secret"
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth)
```

Use the function Cursor for tweepy and api to get 100 tweets containing the word "Boldrin"

```python
boldrinTweets = tweepy.Cursor(api.search_tweets,q = "Boldrin OR boldrin -filter:retweets",lang = "it").items(100)
```

Make a matrix of boldrin tweets 

```python
boldrinTweetsList = [[tweet.created_at, tweet.place, tweet.user.name, tweet.text] for tweet in boldrinTweets]
```

```python
display(boldrinTweetsList)
```

Let's see what happen when we define a dataframe from the above defined list

```python
boldrinDF = pd.DataFrame(boldrinTweetsList)
```

```python
display(boldrinDF)
```

```python
display(pd.DataFrame([range(1,99), boldrinDF[3]]))
```

Let's clean up by removing mentions (i.e. stuff like '@username')

Select the tweet column from the dataframe

```python
tweetColumn = boldrinDF[3]
```

In order to do our work we need to use [regular expression](https://en.wikipedia.org/wiki/Regular_expression). We'll be using 're', a package for doing just that in Python

```python
import re
```

```python
cleanTweetColumn = tweetColumn
```

```python
for i in range(100):
    cleanTweetColumn[i] = re.sub("@[A-Za-z0-9_]+","", tweetColumn[i])
```
