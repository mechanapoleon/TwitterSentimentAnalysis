myGetTweets(keywords, numberOfTweets) is a function that given a text string (keywords) gets the following information for a number of tweets:
-creation date;
-place where it was tweeted;
-username of the creator;
-text of the tweet.

Note: it doesn't get retweets.

```python
def myGetTweets(keywords, numberOfTweets):
    object1 = tweepy.Cursor(api.search_tweets, q = keywords + " -filter:retweets", lang = "it").items(numberOfTweets)
    return [[tweet.created_at, tweet.place, tweet.user.name, tweet.text] for tweet in object1]
```

```python
display(myGetTweets(keywordsSport, 100))
```

```python
def myGetOnlyTweets(keywords, numberOfTweets):
    object1 = tweepy.Cursor(api.search_tweets, q = keywords + " -filter:retweets", lang = "it").items(numberOfTweets)
    object2 = [[tweet.text] for tweet in object1]
    return pd.DataFrame(object2)
```
